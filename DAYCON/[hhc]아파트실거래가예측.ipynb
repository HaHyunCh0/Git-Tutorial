```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
```

```python
train= pd.read_csv('apartment/train.csv', index_col=0) ## id=0부터 시작하게끔 해줌
test = pd.read_csv('apartment/test.csv', index_col=0)
sample_submission = pd.read_csv('apartment/submission.csv')
day_care_center = pd.read_csv('apartment/day_care_center.csv')
park = pd.read_csv('apartment/park.csv', index_col=0)
```

```python
from pandas import Series,DataFrame

df = DataFrame(trainadd)
```

```python
from sklearn import preprocessing

#인코더 생성
#test, train 전처리 
df1 = df
le = preprocessing.LabelEncoder()

df1 = df.drop(["jibun","apt","addr_kr"],axis=1)

year = df['transaction_year_month'].apply( lambda x : str(x)[:4]).astype(int)

df1['transaction_year_month'] = year
df1['transaction_year_month']-=df['year_of_completion']

df1 = df1.drop("year_of_completion",axis=1)

date = le.fit_transform(df['transaction_date'])
df1['transaction_date'] = date
df1
```

```python
train22=mergeTest

train22 = train22.dropna()  #na 인 행 빼기 

train_x = train22.drop(columns=['transaction_real_price'], axis=1)  #값열 빼기
train_y = mergeTest['transaction_real_price'].dropna() #train data만 추출


train_x.shape , train_y.shape
```

```python
rf = RandomForestRegressor(n_estimators = 100)
rf.fit(x_train, y_train)
```
